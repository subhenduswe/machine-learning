```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='blue') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#650bdb') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#650bdb',label='x v x') 

ax.plot(x,x+1,color='purple',label="x vs x+1") 

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#650bdb',lw=3) 

ax.plot(x,x+1,color='purple',linewidth=5) 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#650bdb',lw=10,ls='--') 

ax.plot(x,x+1,color='purple',linewidth=5,linestyle='-.') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='#650bdb',lw=10,ls='--') 

lines = ax.plot(x,x+1,color='purple',linewidth=5) 
lines[0].set_dashes([1,1,1,2,3,5])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='#650bdb',lw=10,ls='--') 

lines = ax.plot(x,x+1,color='purple',linewidth=5) 
lines[0].set_dashes([5,2,10,2])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='#650bdb',lw=10,ls='--') 

ax.plot(x,x+1,color='purple',linewidth=2,marker='o',ls='--') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='purple',linewidth=2,marker='o',ls='--',ms=10) 

ax.plot(x,x+1,color='red',linewidth=2,marker='+',ls='--',markersize=20) 
```
