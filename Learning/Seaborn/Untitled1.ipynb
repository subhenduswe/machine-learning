```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.rugplot(x='salary',data=df)
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.displot(data=df,x='salary')
```

```python
sns.displot(data=df,x='salary',bins=20)
```

```python
#sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20)
```

```python
sns.displot(data=df,x='salary',color='red',edgecolor='blue',
           linewidth=4)
```
