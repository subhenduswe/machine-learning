```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df.head()
```

```python
df = df.set_index("Countries")
```

```python
df
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1))
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5)
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True)
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True)
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,
           cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,
           cmap='viridis',center=40)
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,
           cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,
           cmap='viridis',col_cluster=False)
```
